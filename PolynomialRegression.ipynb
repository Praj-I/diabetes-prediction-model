```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

class PolynomialRegression:
    # Intialize polynomial degree, coefficients, intercept, accuracy score, and features
    def __init__(self, degree = 2):
        self.degree = degree
        self.coefficients = None
        self.intercept = None
        self.r2score = None
        self.poly_features = None
        
    # Trains model. Takes in matrix X, vector y.
    def fit(self, X, y):
        # Transform X into polynomial features
        self.poly_features = PolynomialFeatures(degree = self.degree, include_bias = True)
        X_poly = self.poly_features.fit_transform(X)

        # Compute coefficients using Normal Equation
        self.coefficients_ = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y)

        # Store intercept
        self.intercept_ = self.coefficients_[0]

        # Make predictions
        y_pred = X_poly.dot(self.coefficients_)
        self.y_pred_ = y_pred

        # Store R2 score
        self.r2score_ = 1 - (np.sum((y - y_pred)**2) / np.sum((y - np.mean(y))**2))

    # Outputs model's prediction. Takes in matrix X.
    def predict(self, X):
        X_poly = self.poly_features.transform(X)
        return X_poly.dot(self.coefficients_)
```

```python
# Prep data

diabetes_data = pd.read_csv('diabetes.csv')

traits = [column for column in diabetes_data.select_dtypes(include=[np.number]).columns if column != 'Outcome']

X_poly = diabetes_data[traits].to_numpy()
y_poly = diabetes_data["Outcome"].to_numpy()
```

```python
# Implement Polynomial Regression on diabetes.csv

# Split data into train/test
X_train, X_test, y_train, y_test = train_test_split(X_poly, y_poly, test_size=0.2, random_state=42)

# Train model
polynomialRegression = PolynomialRegression(degree=2)
polynomialRegression.fit(X_poly, y_poly)

# Print coefficients (weights) and R2 score (accuracy)
print(f"Polynomial Regression Coefficients: {polynomialRegression.coefficients_}")
print(f"R² Score: {polynomialRegression.r2score_:.2f}")

# Print predictions
prediction = polynomialRegression.predict(X_test)

print(prediction)
```
