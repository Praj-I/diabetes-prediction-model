```python
import numpy as np
import pandas as pd

class MultipleRegression:

    # Define slopes, bias, model accuracy
    def __init__(self):
        self.coefficients_ = None
        self.intercept_ = None
        self.r2score_ = None

    # Trains model. Takes in matrix X, vector y.
    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y).ravel()
        
        # Add bias column
        n = X.shape[0]
        X_b = np.c_[np.ones((n,1)), X]

        # Compute coefficients using Normal Equation
        self.coefficients_ = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
        self.intercept_ = self.coefficients_[0]

        # Calculate predicted values
        y_pred = X_b.dot(self.coefficients_)

        # Calculate accuracy score
        self.r2score_ = 1 - (np.sum((y - y_pred)**2)/np.sum((y - np.mean(y))**2))

        # Stores predictions for training data
        self.y_pred_ = y_pred
        
    # Outputs model's prediction. Takes in matrix X.
    def predict(self, X):
        # Adds bias column to X
        X_b = np.c_[np.ones((X.shape[0],1)), X]

        # Return predicted values
        return X_b.dot(self.coefficients_)
```

```python
# Prep data

diabetes_data = pd.read_csv('diabetes.csv')

traits = [column for column in diabetes_data.select_dtypes(include=[np.number]).columns if column != 'Outcome']

X_multi = diabetes_data[traits].to_numpy()
y_multi = diabetes_data["Outcome"].to_numpy()
```

```python
# Implement Multiple Linear Regression on diabetes.csv

multipleRegression = MultipleRegression()

# Train the model
multipleRegression.fit(X_multi, y_multi)

# Print the coefficients (weights) and R2 (accuracy) score
print(f"Multiple Linear Regression Coefficients: {multipleRegression.coefficients_}")
print(f"R² Score: {multipleRegression.r2score_:.2f}")

# Print predictions
prediction = multipleRegression.predict(X_multi)
print("Predictions:", prediction)
```
